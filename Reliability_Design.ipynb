# Problem Statement
Write a program to calculate the reliability of a system consisting of three devices connected in series.
# Given Data
* Device 1 has a reliability of 0.9.
* Device 2 has a reliability of 0.8.
* Device 3 has a reliability of 0.5.

```python
# Define reliability values for each device
reliabilities = {
    "Device 1": 0.9,
    "Device 2": 0.8,
    "Device 3": 0.5
}

# Calculate the system reliability for devices connected in series
def series_reliability(reliabilities):
    total_reliability = 1
    for device, reliability in reliabilities.items():
        total_reliability *= reliability
    return total_reliability

# Calculate and print the system reliability
system_reliability = series_reliability(reliabilities)
print(f"The system reliability for devices in series is: {system_reliability:.4f}")
```

# Problem Statement
Write a program to evaluate and visualize the relationship between the cost and reliability of three types of
devices when configured in parallel. The goal is to determine how reliability improves with an increasing
number of devices of the same type while staying within a specified budget.
# Given Data
• Costs of Device Types:
– D1 : 30
– D2 : 15
– D3 : 20
• Reliabilities of Device Types:
– D1 : 0.9
– D2 : 0.8
– D3 : 0.5
• Budget Constraint: 105

```python
import matplotlib.pyplot as plt

# Define costs and reliabilities for each device
costs = [30, 15, 20] # Costs of device types D1, D2, D3
reliabilities = [0.9, 0.8, 0.5] # Reliability of device types D1, D2, D3
budget = 105 # Total budget constraint
# Number of copies allowed for each device type within the budget
max_copies = [
    int(budget // cost) + 1 for cost in costs
]
# Reliability calculation for devices in parallel
def parallel_reliability(r, n):
    return 1 - (1 - r) ** n
# Prepare data for plotting
device_data = {
    "Device 1": {"cost": costs[0], "reliability": reliabilities[0], "max_copies": max_copies[0]},
    "Device 2": {"cost": costs[1], "reliability": reliabilities[1], "max_copies": max_copies[1]},
    "Device 3": {"cost": costs[2], "reliability": reliabilities[2], "max_copies": max_copies[2]},
}
plt.figure(figsize=(10, 6))
# Generate reliability combination graph for each device
for i, (device, data) in enumerate(device_data.items()):
    copies = list(range(1, data["max_copies"] + 1))
    reliability_values = [parallel_reliability(data["reliability"], n) for n in copies]
    cost_values = [data["cost"] * n for n in copies]
    plt.plot(cost_values, reliability_values, marker='o', label=f'{device} Reliability')
# Mark the budget constraint
plt.axvline(x=budget, color='k', linestyle='--', label=f'Budget Constraint (${budget})')
# Labels and Legend
plt.xlabel("Total Cost")
plt.ylabel("Reliability")
plt.title("Reliability Combination Graph for Different Device Configurations")
plt.legend()
plt.grid(True)
plt.show()
```
