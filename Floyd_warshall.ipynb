```python
# Number of vertices
nV = 4
INF = 999

# Algorithm
def floyd(G):
    # Copy the original graph into dist matrix
    dist = list(map(lambda p: list(map(lambda q: q, p)), G))

    # Adding vertices individually as intermediate points
    for r in range(nV):          # For each vertex as intermediate
        for p in range(nV):      # For each source vertex
            for q in range(nV):  # For each destination vertex
                # If the path via r is shorter, update it
                dist[p][q] = min(dist[p][q], dist[p][r] + dist[r][q])

    # Print the final shortest distances
    output_sol(dist)


# Printing the output
def output_sol(dist):
    for p in range(nV):
        for q in range(nV):
            if dist[p][q] == INF:
                print("INF", end=" ")
            else:
                print(dist[p][q], end=" ")
        print(" ")  # Move to the next line after each row


# Input graph matrix (Adjacency Matrix)
G = [
    [0, 5, INF, INF],
    [50, 0, 15, 5],
    [30, INF, 0, 15],
    [15, INF, 5, 0]
]

# Run the Floyd-Warshall Algorithm
floyd(G)
```
